### 4. Hitung Entropy Awal

$$
p_1 = \frac{8}{15} \quad \text{(Citrus hystrix)}, \quad
p_2 = \frac{7}{15} \quad \text{(Citrus Medica)}
$$

$$
Entropy(D) = -\frac{8}{15} \log_2 \frac{8}{15} - \frac{7}{15} \log_2 \frac{7}{15}
$$

$$
\approx -0.533 - 0.463 = \boxed{0.996}
$$

---

### 5. Hitung Entropy Setelah Split
Group Left

$$
Entropy(left) = -1 \cdot \log_2 1 = \boxed{0}

Group Right

$$
Entropy(right) = -1 \cdot \log_2 1 = \boxed{0}
$$

Information Gain:

$$
IG = Entropy(D) - \left( \frac{8}{15} \cdot 0 + \frac{7}{15} \cdot 0 \right)
= \boxed{0.996}
$$

---

### 6. Hitung Gini Index

Gini Awal:

$$
Gini(D) = 1 - \left( \frac{8}{15} \right)^2 - \left( \frac{7}{15} \right)^2
= 1 - 0.284 - 0.218 = \boxed{0.497}
$$

 Gini Setelah Split

* Group Left: purity 100%

  $$
  Gini(left) = 1 - 1^2 = 0
  $$

* Group Right: purity 100%

  $$
  Gini(right) = 1 - 1^2 = 0
  $$

Gini Split:

$$
Gini_{split} = \frac{8}{15} \cdot 0 + \frac{7}{15} \cdot 0 = \boxed{0}
$$

## Visualisasi Decision Tree

Ini adalah code program untuk mencari decision tree sekaligus rules

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree, export_text
from sklearn.preprocessing import LabelEncoder

# === Data ===
data = {
    "Berat (gram)": [300, 280, 320, 290, 310, 270, 330, 80, 75, 85, 70, 90, 78, 82, 88],
    "Diameter (cm)": [10.0, 9.5, 10.5, 9.8, 10.2, 9.3, 10.8, 5.0, 4.8, 5.2, 4.5, 5.5, 4.9, 5.1, 5.4],
    "Kelas": [
        "Citrus Medica", "Citrus Medica", "Citrus Medica", "Citrus Medica", "Citrus Medica",
        "Citrus Medica", "Citrus Medica", "Citrus hystrix", "Citrus hystrix", "Citrus hystrix",
        "Citrus hystrix", "Citrus hystrix", "Citrus hystrix", "Citrus hystrix", "Citrus hystrix"
    ]
}
df = pd.DataFrame(data)

# === Label Encoding ===
le = LabelEncoder()
df['Label'] = le.fit_transform(df['Kelas'])  # Citrus hystrix=0, Citrus Medica=1

# === Fitur dan Target ===
X = df[["Berat (gram)", "Diameter (cm)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === 1. Tampilkan Pohon Keputusan ===
print("\n=== Pohon Keputusan (Decision Tree) ===")
tree_rules = export_text(clf, feature_names=list(X.columns))
print(tree_rules)

```

## Visualisasi Splitting

```python
# === Fitur dan Target ===
X = df[["Berat (gram)", "Diameter (cm)"]]
y = df["Label"]

# === Train Decision Tree ===
clf = DecisionTreeClassifier(criterion="gini", max_depth=3, random_state=42)
clf.fit(X, y)

# === Uji 2 Data Baru ===
uji_baru = pd.DataFrame({
    'Berat (gram)': [315, 72],
    'Diameter (cm)': [10.1, 4.6]
})

prediksi = clf.predict(uji_baru)
hasil = le.inverse_transform(prediksi)

print("\n=== Prediksi 2 Data Baru ===")
for i, (berat, diameter) in enumerate(zip(uji_baru['Berat (gram)'], uji_baru['Diameter (cm)'])):
    print(f"Data {i+1}: Berat={berat}g, Diameter={diameter}cm → Prediksi: {hasil[i]}")

# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 1.0

x_min, x_max = X['Berat (gram)'].min() - 10, X['Berat (gram)'].max() + 10
y_min, y_max = X['Diameter (cm)'].min() - 0.5, X['Diameter (cm)'].max() + 0.5

xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, 0.1))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Berat (gram)', 'Diameter (cm)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Berat (gram)'], X.iloc[idx]['Diameter (cm)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

plt.xlabel("Berat (gram)")
plt.ylabel("Diameter (cm)")
plt.title("Decision Boundary (Berat vs Diameter)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi Splitting Setelah ada 2 Data Baru

```python
# === Visualisasi Decision Boundary ===
print("\n=== Visualisasi Decision Boundary ===")
plot_colors = "ry"
plot_step = 1.0

# Sesuaikan rentang berdasarkan fitur "Berat (gram)" dan "Diameter (cm)"
x_min, x_max = X['Berat (gram)'].min() - 10, X['Berat (gram)'].max() + 10
y_min, y_max = X['Diameter (cm)'].min() - 0.5, X['Diameter (cm)'].max() + 0.5

# Buat grid untuk seluruh area
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step),
                     np.arange(y_min, y_max, 0.1))

grid = np.c_[xx.ravel(), yy.ravel()]
grid_df = pd.DataFrame(grid, columns=['Berat (gram)', 'Diameter (cm)'])
Z = clf.predict(grid_df)
Z = Z.reshape(xx.shape)

plt.figure(figsize=(8, 6))
plt.contourf(xx, yy, Z, cmap=plt.cm.RdYlBu, alpha=0.6)

# Plot titik data training
for label, color in zip([0, 1], plot_colors):
    idx = np.where(y == label)
    plt.scatter(X.iloc[idx]['Berat (gram)'], X.iloc[idx]['Diameter (cm)'],
                c=color, edgecolor='k', label=le.inverse_transform([label])[0])

# Plot titik data uji jika ada (opsional)
plt.scatter(uji_baru['Berat (gram)'], uji_baru['Diameter (cm)'],
            c='black', marker='X', s=100, label='Data Uji')

plt.xlabel("Berat (gram)")
plt.ylabel("Diameter (cm)")
plt.title("Decision Boundary (Berat vs Diameter)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Visualisasi Decision Tree dalam bentuk gambar

```python
# === Visualisasi Struktur Pohon ===
plt.figure(figsize=(12, 3))
plot_tree(
    clf,
    feature_names=["Berat (gram)", "Diameter (cm)"],  # Ganti nama fitur
    class_names=le.classes_,                          # Tetap gunakan label string
    filled=True,
    rounded=True,
    impurity=True                                     # Menampilkan nilai Gini
)
plt.title("Decision Tree untuk Citrus Medica vs Citrus hystrix")
plt.tight_layout()
plt.show()
```
